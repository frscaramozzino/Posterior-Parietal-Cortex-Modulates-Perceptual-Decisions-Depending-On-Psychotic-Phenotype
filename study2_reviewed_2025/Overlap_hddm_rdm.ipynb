```python
import arviz as az
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pymc3 as pm
import theano
import xarray as xr
from statistics import NormalDist

print(f"Running on PyMC3 v{pm.__version__}")
```

```python
%config InlineBackend.figure_format = 'retina'
az.style.use("arviz-darkgrid")
```

```python
#ovl for vat_tms 
vat_tms_stats=pd.read_csv('vat_tms_traces.csv')
vat_tms_stats
```

```python
v_mu_sham=vat_tms_stats['v(0)'].mean()
v_sigma_sham=vat_tms_stats['v(0)'].std()

v_mu_tms=vat_tms_stats['v(1)'].mean()
v_sigma_tms=vat_tms_stats['v(1)'].std()

print ("Beta TMS =/0") 
ovl_v_tms_sham=NormalDist(mu=v_mu_tms, sigma=v_sigma_tms).overlap(NormalDist(mu=v_mu_sham, sigma=v_sigma_sham))
print (ovl_v_tms_sham)
```

```python
a_mu_sham=vat_tms_stats['a(0)'].mean()
a_sigma_sham=vat_tms_stats['a(0)'].std()

a_mu_tms=vat_tms_stats['a(1)'].mean()
a_sigma_tms=vat_tms_stats['a(1)'].std()

print ("Beta TMS =/0") 
ovl_a_tms_sham=NormalDist(mu=a_mu_tms, sigma=a_sigma_tms).overlap(NormalDist(mu=a_mu_sham, sigma=a_sigma_sham))
print (ovl_a_tms_sham)
```

```python
t_mu_sham=vat_tms_stats['t(0)'].mean()
t_sigma_sham=vat_tms_stats['t(0)'].std()

t_mu_tms=vat_tms_stats['t(1)'].mean()
t_sigma_tms=vat_tms_stats['t(1)'].std()

print ("Beta TMS =/0") 
ovl_t_tms_sham=NormalDist(mu=t_mu_tms, sigma=t_sigma_tms).overlap(NormalDist(mu=t_mu_sham, sigma=t_sigma_sham))
print (ovl_t_tms_sham)
```
